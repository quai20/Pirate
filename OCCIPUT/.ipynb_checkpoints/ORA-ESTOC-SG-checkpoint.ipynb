## LOAD DATASETS

```python
#LOAD DATASETS & VARIABLE TO INTERP
#ORAS4
ORA_T=xr.open_mfdataset('/home5/pharos/REFERENCE_DATA/OCEAN_REP/ORAS4/DATA/monthly_1x1/thetao_oras4_1m_200[5,6,7,8]_grid_1x1.nc')
ORA_S=xr.open_mfdataset('/home5/pharos/REFERENCE_DATA/OCEAN_REP/ORAS4/DATA/monthly_1x1/so_oras4_1m_200[5,6,7,8]_grid_1x1.nc')
ORA_T.thetao.load()
ORA_S.so.load()
ORA_S
```

```python
#ESTOC
ESTOC_T=xr.open_mfdataset('/home5/pharos/REFERENCE_DATA/OCEAN_REP/ESTOC/DATA/200[5,6,7,8]/k7oda_tmp_*.nc',decode_times=False)
ESTOC_S=xr.open_mfdataset('/home5/pharos/REFERENCE_DATA/OCEAN_REP/ESTOC/DATA/200[5,6,7,8]/k7oda_sal_*.nc',decode_times=False)
#Handling date
date=np.empty(len(ESTOC_T.TIME.values),dtype='datetime64[D]')
for i in range(len(ESTOC_T.TIME.values)):
    Y=str(np.datetime64('1900')+ESTOC_T.TIME[i].values/12)
    Mi=(ESTOC_T.TIME[i].values % 12)+1
    if Mi<10:
        M='0'+str(Mi)
    else:
        M=str(Mi)
    date[i]=np.datetime64(Y+'-'+M+'-01','D')
ESTOC_T['TIME']=date
ESTOC_S['TIME']=date

ESTOC_T.TMP.load()
ESTOC_S.SAL.load()
ESTOC_S
```

```python
# SG-SCRIPPS
SG_T=xr.open_mfdataset('/home5/pharos/REFERENCE_DATA/OCEAN_REP/SCRIPPS/200[5,6,7,8]/*_Temp_*.nc',decode_times=False)
SG_S=xr.open_mfdataset('/home5/pharos/REFERENCE_DATA/OCEAN_REP/SCRIPPS/200[5,6,7,8]/*_PSal_*.nc',decode_times=False)

#Handling date
date=np.empty(len(SG_T.TIME.values),dtype='datetime64[D]')
for i in range(len(SG_T.TIME.values)):
    Y=str(np.datetime64('2004')+int(SG_T.TIME[i].values/12))
    Mi=int((SG_T.TIME[i].values-0.5) % 12)+1
    if Mi<10:
        M='0'+str(Mi)
    else:
        M=str(Mi)
    date[i]=np.datetime64(Y+'-'+M+'-15','D')
SG_T['TIME']=date
SG_S['TIME']=date

SG_T.ARGO_TEMPERATURE_MEAN.load()
SG_S.ARGO_SALINITY_MEAN.load()
SG_T
```

## INTERPOLATION

```python
#INIT
theta_ORA=np.zeros_like(DAT.POTM_OBS.values)*np.nan
psal_ORA=np.zeros_like(DAT.PSAL_OBS.values)*np.nan
theta_ESTOC=np.zeros_like(DAT.POTM_OBS.values)*np.nan
psal_ESTOC=np.zeros_like(DAT.PSAL_OBS.values)*np.nan
theta_SG=np.zeros_like(DAT.POTM_OBS.values)*np.nan
psal_SG=np.zeros_like(DAT.PSAL_OBS.values)*np.nan

for i in DAT.N_OBS.values:    
    tio=pd.to_datetime(DAT.JULD[i].values,unit='D',origin=pd.to_datetime('1950/1/1'))
    lao=DAT.LATITUDE[i].values
    loo=DAT.LONGITUDE[i].values
    zio=DAT.STANDARD_LEVELS.values
    
    #ORAS4
    zio[0]=zio[0]-0.03  #SHIFT FIRST LEVEL FOR INTERPOLATION SUCCESS (CHEATING I KNOW...)
    INT_t=ORA_T.thetao.interp(time=tio, lat=lao, lon=np.mod(loo+360,360), depth=-zio)
    INT_s=ORA_S.so.interp(time=tio, lat=lao, lon=np.mod(loo+360,360), depth=-zio)
    theta_ORA[i,:]=INT_t.values        
    psal_ORA[i,:]=INT_s.values
    
    #ESTOC    
    INX_t=ESTOC_T.TMP.interp(TIME=tio, LATITUDE=lao, LONGITUDE=np.mod(loo+360,360), DEPTH=-zio)
    INX_s=ESTOC_S.SAL.interp(TIME=tio, LATITUDE=lao, LONGITUDE=np.mod(loo+360,360), DEPTH=-zio)
    theta_ESTOC[i,:]=INX_t.values        
    psal_ESTOC[i,:]=INX_s.values    
    
    #SG-SCRIPPS   
    pio=gsw.p_from_z(zio,lao)
    INY_t=SG_T.ARGO_TEMPERATURE_MEAN.interp(TIME=tio, LATITUDE=lao, LONGITUDE=np.mod(loo+360,360), PRESSURE=pio)
    INY_s=SG_S.ARGO_SALINITY_MEAN.interp(TIME=tio, LATITUDE=lao, LONGITUDE=np.mod(loo+360,360), PRESSURE=pio)
    SA=gsw.SA_from_SP(INY_s.values, pio, loo, lao)
    POTM=gsw.pt0_from_t(SA,INY_t.values,pio)
    theta_SG[i,:]=POTM
    psal_SG[i,:]=INY_s.values         
    
```

```python
%matplotlib notebook
inz=5
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.plot(DAT.POTM_OBS[inz,:].values,zio,linewidth=4,label='OBS')
plt.plot(DAT.POTM_OTH[0,inz,:].values,zio,label='ISAS')
plt.plot(DAT.POTM_OTH[1,inz,:].values,zio,label='ECCO')
plt.plot(theta_ORA[inz,:],zio,label='ORAS4')
plt.plot(theta_ESTOC[inz,:],zio,label='ESTOC')
plt.plot(theta_SG[inz,:],zio,label='SCRIPPS')
plt.fill_betweenx(zio,DAT.POTM_Hx[:,inz,:].min('N_MEMBER'),DAT.POTM_Hx[:,inz,:].max('N_MEMBER'),
                  facecolor='gray', alpha=0.3, label='OCCIPUT')
plt.grid(linestyle=':')
plt.legend()
plt.subplot(1,2,2)
plt.plot(DAT.PSAL_OBS[inz,:].values,zio,linewidth=4,label='OBS')
plt.plot(DAT.PSAL_OTH[0,inz,:].values,zio,label='ISAS')
plt.plot(DAT.PSAL_OTH[1,inz,:].values,zio,label='ECCO')
plt.plot(psal_ORA[inz,:],zio,label='ORAS4')
plt.plot(psal_ESTOC[inz,:],zio,label='ESTOC')
plt.plot(psal_SG[inz,:],zio,label='SCRIPPS')
plt.fill_betweenx(zio,DAT.PSAL_Hx[:,inz,:].min('N_MEMBER'),DAT.PSAL_Hx[:,inz,:].max('N_MEMBER'),
                  facecolor='gray', alpha=0.3, label='OCCIPUT')
plt.grid(linestyle=':')
plt.show()
```
