```python
# -*- coding: utf-8 -*
import xarray as xr
import pandas as pd
import numpy as np
import glob
import os
import datetime
import matplotlib.pyplot as plt
from matplotlib import cm
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from pyxpcm import pcm

DAT=xr.open_dataset('/export/home1/openbar/ISAS_data_allNA.nc')
```

```python
# DAT=DAT.where((DAT.LAT>40),drop=True)
# DAT=DAT.squeeze()

ntemp=np.reshape(np.transpose(DAT.TEMP.values,[1,0,2,3]),[152,202*85*159])

ARRAY=xr.Dataset({'temp':(['DEPTH','N_PROF'],ntemp[:102,:])},
                 coords={'DEPTH':DAT.DEPT[:102,0].values})      
```

```python
H1=ARRAY['DEPTH'].where(ARRAY['temp'].notnull()).max(dim='DEPTH')
H2=ARRAY['DEPTH'].where(ARRAY['temp'].notnull()).min(dim='DEPTH')
ARRAY = ARRAY.where((H1>=1000)&(H2<=10), drop=True)
ARRAY
```

```python
#BIC SIMULATION TO DECIDE HOW MANY CLUSTERS WE USE
kmax = 19
ntry = 10
BIC = np.zeros([ntry,kmax])
n=50

for lt in range(ntry):    
    if(np.mod(lt,10)==0):
        print lt
    randsel=np.random.choice(range(len(ARRAY.N_PROF)),n,False)
    subtrain=ARRAY.where(ARRAY.N_PROF.isin(randsel),drop=True)    
    for k in range(kmax):
        m = pcm(K=k+1, feature_axis=ARRAY.DEPTH.values, feature_name='temperature')
        A=m.fit(subtrain, feature={'temperature': 'temp'})        
        llih=m.score(subtrain, feature={'temperature': 'temp'})                 
        Xn=m.preprocessing(subtrain,feature={'temperature':'temp'})
        #D=Dimension de la decomposition verticale des profiles (EOF)
        D = Xn.shape[1]
        Nf = (k+1)-1 + (k+1)*D + (k+1)*D*(D+1)/2
        BIC[lt,k] = -2*n*llih + Nf*np.log(n) #    
        #print llih*n,Nf,D
        
plt.figure(figsize=(10,5))
h1=plt.errorbar(np.arange(kmax)+1, np.mean(BIC,axis=0), yerr=np.std(BIC,axis=0),fmt='k-.',linewidth=2,
                ecolor='k',elinewidth=1,capsize=2,capthick=1)
plt.plot(np.arange(kmax)+1, np.mean(BIC,axis=0)+np.std(BIC,axis=0),linewidth=0.5,color=[0.7]*3)
plt.plot(np.arange(kmax)+1, np.mean(BIC,axis=0)-np.std(BIC,axis=0),linewidth=0.5,color=[0.7]*3)
plt.gca().grid(linestyle=':')
# plt.ylim(2600,4000)
plt.legend(h1,['Mean BIC score - '+str(ntry)+' runs'])
plt.show()        
```
